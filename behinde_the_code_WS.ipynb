## Criando um método 

Depois que eu vi que meus testes estavam dando certo, Eu irei fazer um método que pegar uma lista de url que devolve jsons com os conteudos  
desejados 

Segue o código:


````python
def ted_talk_scraping(urls):
    from bs4 import BeautifulSoup
    import requests 
    import json
    
    title_json = 1
    for url in urls:
        
        res =  requests.get(url)
        res.encoding = 'utf-8'
        soup = BeautifulSoup(res.text,'html.parser')
        
        title = soup.find("meta",  itemprop="name")["content"]
        author = soup.find("span",  itemprop="author").find('meta', itemprop="name")["content"]


        a = soup.find("div",["Grid__cell", "flx-s:1"]).find_all("p")
        text = ''
        for i in a:
            text += (i.text+'\n')
            
        body  = text.replace('\t', '').replace('(Risos)', '\n').replace('(Aplausos)', '\n').replace('(Vivas)', '\n')
        
        x = {"author": author, "body": body, "title": title , "type": "video", "url": url}
        
        with open(str(title_json) +'.json', 'w') as json_file:
            json.dump(x, json_file, indent = 3, ensure_ascii = False)
            
        title_json += 1
````

````python
ted_talk_scraping(urls)
````

````python
def olhar_digital_scraping(urls):
    from bs4 import BeautifulSoup
    import requests 
    import json
    
    title_json = 8
    for url in urls:
        
        res =  requests.get(url)
        soup = BeautifulSoup(res.text,'html.parser')
        
        title = soup.find('meta', property="og:title")['content']
        author = soup.find('meta', property="article:author")['content']


        y = soup.find('div','mat-txt').find_all('p')
        
        body = ''
        for x in y:
            body += (x.text + '\n')
                    
        x = {"author": author, "body": body, "title": title , "type": "article", "url": url}
        
        with open(str(title_json) +'.json', 'w') as json_file:
            json.dump(x, json_file, indent = 3, ensure_ascii = False)
            
        title_json += 1
````
